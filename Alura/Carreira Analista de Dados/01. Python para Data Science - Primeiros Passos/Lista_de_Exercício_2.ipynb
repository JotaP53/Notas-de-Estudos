1) Escreva um programa que peça à pessoa usuária para fornecer dois números e exibir o número maior.

```python
num1 = int(input('Digite um número: '))
num2 = int(input('Digite outro número: '))

if num1 > num2:
    print(f'O número maior é: {num1}.')
else:
    print(f'O número maior é {num2}.')
```

2) Escreva um programa que solicite o percentual de crescimento de produção de uma empresa e informe se houve um crescimento (porcentagem positiva) ou decrescimento (porcentagem negativa).

```python
cresc_prod1 = float(input('Digite o percentual 1 '))
cresc_prod2 = float(input('Digite o percentual 2 '))

if cresc_prod2 > cresc_prod1:
    print(f'Houve um crescimento de {cresc_prod2 - cresc_prod1}%.')
elif cresc_prod2 == cresc_prod1:
    print(f'A produção se manteve estável.')
else:
    print(f'Houve um decrescimento de {cresc_prod2 - cresc_prod1}%.')
```

3) Escreva um programa que determine se uma letra fornecida pela pessoa usuária é uma vogal ou consoante.

```python
letra = input('Digite uma letra: ')
vogais = 'a, e, i, o, u, A, E, I, O, U'

if letra in vogais:
    print(f'{letra} é uma vogal.')
else:
    print(f'{letra} é uma  consoante.')
```

4) Escreva um programa que leia valores médios de preços de um modelo de carro por 3 anos consecutivos e exiba o valor mais alto e mais baixo entre esses três anos.

```python
preco_medio_1 = float(input('Digite o preço do ano 1: '))
preco_medio_2 = float(input('Digite o preço do ano 2: '))
preco_medio_3 = float(input('Digite o preço do ano 3: '))

print(f'O valor do ano 1 é: {preco_medio_1} reais.')
print(f'O valor do ano 2 é: {preco_medio_2} reais.')
print(f'O valor do ano 3 é: {preco_medio_3} reais.')

if preco_medio_1 > preco_medio_2:
    if preco_medio_1 > preco_medio_3:
        print(f'O preço mais alto é o do ano 1.')
elif preco_medio_2 > preco_medio_1:
    if preco_medio_2 > preco_medio_3:
        print(f'O preço mais alto é o do ano 2.')
else:
    print(f'O preço mais alto é o do ano 3.')
```

5) Escreva um programa que pergunte sobre o preço de três produtos e indique qual é o produto mais barato para comprar.

```python
preco_produto_1 = float(input('Digite o preço do produto 1: '))
preco_produto_2 = float(input('Digite o preço do produto 2: '))
preco_produto_3 = float(input('Digite o preço do produto 3: '))

print(f'O valor do produto 1 é: {preco_produto_1} reais.')
print(f'O valor do produto 2 é: {preco_produto_2} reais.')
print(f'O valor do produto 3 é: {preco_produto_3} reais.')

if preco_produto_1 < preco_produto_2:
    if preco_produto_1 < preco_produto_3:
        print('O produto mais barato é o produto 1')
elif preco_produto_2 < preco_produto_1:
    if preco_produto_2 < preco_produto_3:
        print('O produto mais barato é o produto 2')
else:
    print('O produto mais barato é o produto 3')
```

6) Escreva um programa que leia três números e os exiba em ordem decrescente.

```python
a = float(input('Digite o número 1: '))
b = float(input('Digite o número 2: '))
c = float(input('Digite o número 3: '))

print(f'O número 1 é: {a}.')
print(f'O número 2 é: {b}.')
print(f'O número 3 é: {c}.')

if a < b < c:
    print(f'{c, b, a}')
elif a < c < b:
    print(f'{b, c, a}')
elif b < c < a:
    print(f'{a, c, b}')
else:
    print(f'{a, b, c}')
```

8) Escreva um programa que peça um número inteiro à pessoa usuária e determine se ele é par ou ímpar. Dica: Você pode utilizar o operador módulo `%`.

```python
num = int(input('Digite um número:'))

if num % 2 == 0:
    print(f'O número {num} é par.')
else:
    print(f'O número {num} é ímpar.')
```
