# Tipos de Variáveis em Python

Uma variável é um objeto, afinal tudo é um objeto em Python

int -> inteiro<br>
string -> texto<br>
float -> números com casas decimais (ponto flutuante)<br>
bool ou boolean -> True ou False<br>

Obs: Variáveis em Python não são declaradas explicitamente, mas isso não significa que você deve ficar mudando o tempo todo

Misturando variáveis

```python
faturamento = 1500
custo = 500
lucro = faturamento - custo
print('faturamento da loja foi ' + str(faturamento) + ' reais')
print('O custo da loja foi ' + str(custo) + ' reais')
print('O luvro da loja foi ' + str(lucro) + ' reais')

# concantenando tudo em um print
print('O faturamento da loja foi ' + str(faturamento) + ' reais, o custo foi ' + str(custo) + ' reais e o lucro foi ' + str(lucro) + ' reais.')
```

Método Format

```python
faturamento = 1500
custo = 500
lucro = faturamento - custo

# sem método format
print('O faturamento da loja foi ' + str(faturamento) + ' reais, o custo foi ' + str(custo) + ' reais e o lucro foi ' + str(lucro) + ' reais.')

# com método format
print('O faturamento da loja foi {} reais, o custo foi {} reais e o lucro foi {} reais.'.format(faturamento, custo, lucro))
```

Método F-String

```python
faturamento = 1500
custo = 500
lucro = faturamento - custo

# com método format
print('O faturamento da loja foi {} reais, o custo foi {} reais e o lucro foi {} reais.'.format(faturamento, custo, lucro))

# com método f-string
print(f'O faturamento da loja foi {faturamento} reais, o custo foi {custo} reais e o lucro foi {lucro} reais.')
```

Comparando as três formas (perceba a diminuição da necessidade de escrever muito)

```python
faturamento = 1500
custo = 500
lucro = faturamento - custo

# sem método format
print('O faturamento da loja foi ' + str(faturamento) + ' reais, o custo foi ' + str(custo) + ' reais e o lucro foi ' + str(lucro) + ' reais.')

# com método format
print('O faturamento da loja foi {} reais, o custo foi {} reais e o lucro foi {} reais.'.format(faturamento, custo, lucro))

# com método f-string
print(f'O faturamento da loja foi {faturamento} reais, o custo foi {custo} reais e o lucro foi {lucro} reais.')
```
